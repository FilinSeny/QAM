```python
import matplotlib.pyplot as plt
import numpy as np

def plot_functions():
    functions = []
    for _ in range(3):
        function_data = input().split()
        x_values = [float(function_data[i]) for i in range(0, len(function_data), 2)]
        y_values = [float(function_data[i]) for i in range(1, len(function_data), 2)]
        functions.append((x_values, y_values))

    plt.figure(figsize=(10, 6))
    for i, (x, y) in enumerate(functions, start=1):
        plt.plot(x, y, label=f'QAM {2**(2*i)}')

    plt.xlabel('SNR dB')
    plt.ylabel('BER %')
    plt.title('BER / SNP for QAMs')
    plt.legend()
    plt.grid(True)
    plt.yscale("log")
    plt.show()
    

if __name__ == "__main__":
    plot_functions()
```
